## .set_index() and .reset_index() Methods

```python
bond = pd.read_csv('jamesbond.csv')
bond.head(5)
```

```python
#  設置索引，索引並不會真的被設置，除非你加入參數inplace=True
bond.set_index('Film', inplace=True)
bond.head()
```

```python
#  如果這欄不要了，那可以透過drop=True來刪除(記得搭配inplace)
bond.reset_index(drop=True)
```

```python
#  即使你設置了inplace=True,還是可以透過reset_index來還原
bond.reset_index(inplace=True)
bond.head()
```

```python
#  在設置索引的情況下如果又設置索引的話，原欄會消失
#  所以記得要先reset_index再做set
bond.set_index('Film', inplace=True)
bond.head(5)
```

```python
bond.reset_index(inplace=True)
bond.set_index('Year').head()
```

## Retrieve Rows by index Label with .loc[]

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(5)
```

```python
#  設置好index並且排序之後，就可以透過loc來尋找，排序之後速度會更快
#  回傳的格式是series，欄位即為索引，與相關值一併呈現
bond.loc['Goldfinger']
```

```python
#  當搜尋的資料不止一筆的時候，它就會回傳dataframe的格式
bond.loc['Casino Royale']
```

```python
#  也可以搜尋區間，注意到，包含了moonraker，記得python的slice特性嗎?是m-1
#  但pandas是包括了
bond.loc['Diamonds Are Forever':'Moonraker']
```

```python
#  也可以利用list格式來做多條件的資料搜尋
bond.loc[['Moonraker', 'Octopussy']]
```

```python
#  當加入一個不存在的資料搜尋時，會整排呈現null
bond.loc[['Moonraker', 'Octopussy', 'ABC']]
```

```python
#  這時候可以用典型的python語法查看是否存在該index
'ABC' in bond.index
```

## Set Multiple Values in DataFrame

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#  標準來說，我們要對某個欄位做值調整，可能會想到用遮罩的方式來處理
mask = bond['Actor'] =='Sean Connery'
bond[mask]
```

```python
#  然後再利用遮罩之後來調整，但這會造成失敗，為什麼?
#  不要忘了，我們所做的每個操作都不是對原dataframe做操作，而是產生一個新的副本
bond[mask]=='Shaoe.Chen'
```

```python
#  除非你另外用一個變數去接然後再來調整
df2 = bond[mask]
df2['Actor'] = 'Shaoe.Chen'
df2
```

```python
#  所以，我們要調整一下作法，利用遮罩取到相對應的index與欄位再做賦值調整
bond.loc[mask,'Actor'] = 'Shaoe.Chen'
bond[bond['Actor']=='Shaoe.Chen']
```

## Filtering with the where Method

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#  一般來說，我們要搜尋某個條件的話，會先產生一個遮罩，然後去呈現
#  只是這樣子的話只會回傳有的資料
mask = bond['Actor'] =='Sean Connery'
bond[mask]
```

```python
#  可以利用where，符合條件的會有值，未能符合的就整列回傳null
#  這樣你可以對資料集有較佳的觀察
bond.where(mask)
```

```python
#  也可以利用&(and)與|(or)來做多條件的選擇
mask2 = bond['Box Office']>800
bond.where(mask & mask2)
```

```python
bond.where(mask | mask2)
```

## .query() Method

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#  當我們使用query的時候，參數是字串
#  適用於欄位名稱沒有空格的情況，像Box Office就不行
#  先用串列生成式調整欄位名稱
bond.columns = [column_name.replace(' ','_') for column_name in bond.columns]
```

```python
#  可以查詢文字類型
bond.query('Actor == "Roger Moore"')
```

```python
#  可以查詢數值類型
bond.query('Budget==54.5')
```

```python
#  可以用and多條件
bond.query('Budget==54.5 and Actor=="Roger Moore"')
```

```python
#  可以用in做欄位多條件
#  反過來做not in也可以
bond.query('Actor in ["Timothy Dalton","George Lazenby"]')
```

## .apply() Method on Single Columns

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#  apply應用於自定義函數調整欄位
def convery_salary(salary):
    return str(salary) + 'Millons'
```

```python
bond['Box Office'] = bond['Box Office'].apply(convery_salary)
```

```python
bond.head()
```

```python
#  可以利用迴圈來批次調整
columns = ['Budget', 'Bond Actor Salary']
for col in columns:
    bond[col] = bond[col].apply(convery_salary)
    
bond.head()
```

## .apply() Method with Row Values

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#  基本上觀念跟asp.net的gridview一樣，每一row都可以再細到cell
def movie_val(row):
    actor = row[1]
    budget = row[4]
    if actor == 'Pierce Brosnan':
        return 'Very Good'
    else:
        return 'Other'
```

```python
#  我們判斷了row>cell,所以如果你以axis=index的角度會異常
bond.apply(movie_val)
```

```python
bond.apply(movie_val, axis='columns')
```

```python
#  增加新的欄位保存你的計算
bond['NewCloumn']=bond.apply(movie_val, axis='columns')
```

```python
bond.head()
```

## .copy() Method

```python
bond = pd.read_csv('jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#  我們透過變數來接Director 這個series
#  會發現，連同index都copy過來了
directors = bond['Director']
directors.head()
```

```python
#  然後我們調整這個series的欄位值
#  這會得到一個異常訊息
directors['A View to a Kill'] = 'Marty Chen'
```

```python
#  但是資料確實有調整
directors.head()
```

```python
#  然後我們回頭看dataframe
#  發現也異動了，這是因為雖然切片出去了，但還是參照了dataframe
bond.head()
```

```python
#  我們透過copy來處理欄位切片
directors2 = bond['Director'].copy()
directors2.head()
```

```python
directors2['A View to a Kill']='Super Marty'
directors2.head()
```

```python
#  透過copy來切片，就不會有參照問題的產生了
bond.head()
```
