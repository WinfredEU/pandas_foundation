## The pd.concat() Method Part1

```python
weak1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
weak2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
#  利用pandas的concat來堆疊資料，但會發現一件事，為何資料應該有500筆
#  但是索引卻只出現249(250筆)?
#  這跟參數ignore_index有關
pd.concat([weak1, weak2])
```

```python
len(pd.concat([weak1, weak2]))
```

```python
#  主要問題在於，兩個dataframe有相同的index，都是由0-249，所以覆蓋了
#  當我們透過參數ingore_index設置為True的時候，就可以明確的顯示了。
pd.concat([weak1, weak2], ignore_index=True)
```

```python
#  當column名稱不同的時候，pandas不會知道你要什麼
#  所以就會把它就這樣串聯起來而以，該null的給null
pd.concat([weak1, foods])
```

## Inner join Part1

```python
weak1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
weak2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
#  注意到參數中有一個how，這是決定join的方式，inner、left、right
#  on的部份必需欄位的名稱相同
#  回傳的是一個新的dataframe，並且自動的將我們的欄位做了區隔，避免混亂
#  當然我們會發現有客戶怎麼出現兩筆了，這是因為該客戶在一週內買了多件產品
weak1.merge(weak2, how='inner', on='Customer ID')
```

```python
weak2[weak2['Customer ID']==155]
```

```python
#  後綴的部份當然也是可以調整，預設_x、_y可以從參數說明看的出來
weak1.merge(weak2, how='inner', on='Customer ID', suffixes=['_weak1', '_weak2']).head()
```

## Outer joins

```python
weak1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
weak2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
#  這次嚐試outer join
weak1.merge(weak2, how='outer', on='Customer ID', suffixes=['_weak1', '_weak2'])
```

```python
#  利用indicator來顯示資料join的來源
#  回傳資料可以明確看到，有來自both，來自left only、right only
weak1.merge(weak2, how='outer', on='Customer ID'
            , suffixes=['_weak1', '_weak2'], indicator=True)
```

```python
#  那麼indicator的用處在那?透過欄位我們可以快速的理解資料集
#  我們看到存在第二週的資料有197筆，單存的第一週的有195..
merged=weak1.merge(weak2, how='outer', on='Customer ID'
            , suffixes=['_weak1', '_weak2'], indicator=True)
merged['_merge'].value_counts()
```

```python
#  也可以建立遮罩來快速取得需求資料
mark = merged['_merge'].isin(['left_only', 'right_only'])
merged[mark]
```

## left_on and right_on

```python
weak1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
weak2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
#  在之前的案例中，我們透過on來設置join的欄位，但這都是建立在欄位名稱相同的情況
#  如果是在不相同的情況下，該如何處理?left_on and right_on
#  weak2是customer id
weak2.head()
```

```python
#  客戶主檔內是id
customers.head()
```

```python
#  利用left_on and right_on來執行join
weak2.merge(customers, how='inner', left_on='Customer ID', right_on='ID')
```

```python
#  如果用on做join?
weak2.merge(customers, how='inner', on='Customer ID')
```

```python
#  join之後，如果有不需要的欄位再執行drop
weak2.merge(customers, how='inner', left_on='Customer ID', right_on='ID').drop('ID', axis='columns').head()
```

## Merging b Indexes with  left_index and right_index

```python
weak1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
weak2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
#  加入id做索引
customers = pd.read_csv('Restaurant - Customers.csv', index_col='ID')
#  加入food_id做索引
foods = pd.read_csv('Restaurant - Foods.csv', index_col='Food ID')
```

```python
#  利用索引來做join的欄位，設置right_index=True，代表以id做鍵值 
#  使用索引進行合併的優點是不會出現重覆的列
weak1.merge(customers, how='left', left_on='Customer ID', right_index=True)
```

```python
sales = weak1.merge(customers, how='left', left_on='Customer ID'
                    , right_index=True)
sales = sales.merge(foods, how='left', left_on='Food ID', right_index=True)
sales.head()
```

```python
#  練習其它的join
weak1.merge(weak2, how='left', left_index=True, right_index=True, suffixes=['_weak1','_weak2'])
```
